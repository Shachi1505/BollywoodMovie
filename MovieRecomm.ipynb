```python
import numpy as np
import pandas as pd
df = pd.read_csv('movies_data.csv')
df
```

```python
#Removing duratio, actor 3 -> data cleaning
df = df[['Name', 'Year', 'Genre', 'Rating',	'Director', 'Actor 1']]
df
```

```python
import re #cleaning data
def clean_title(Name):
    return re.sub("[^a-zA-Z0-9]", "", Name)
df["clean_title"] = df["Name"].apply(clean_title)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range = (1,2))

tfidf = vectorizer.fit_transform(df["clean_title"])

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

Name = "Aur Pyaar Ho Gaya"
Name = clean_title(Name)
query_vec = vectorizer.transform([Name])
similarity = cosine_similarity(query_vec, tfidf).flatten()
indices = np.argpartition(similarity, -5) [-5:]
results = df.iloc[indices]
results
```
